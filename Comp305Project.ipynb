```python
try:
  input_file = open("small-testdata.in","r")
  output_file = open("small-testdata.out","r") 
except: 
  print("Problem occured while opening the files!")

totalCases = int(input_file.readline())
print("Number of total cases : ",totalCases)

wholeCasesList = []
for i in range(totalCases):
  aCase = []
  totalPerson = int(input_file.readline())
  for j in range(totalPerson):
    aLine = input_file.readline()
    values  = aLine.split(" ")
    for k in range(len(values)): #This is 0 to 5 always
      aCase.append((int(values[k]),j))
  wholeCasesList.append(aCase)
  print("Case Number {} ****  Number of people : {} ----> {}".format(i+1,totalPerson,aCase))
```

```python
def merge(arr, l, m, r):
    n1 = m - l + 1
    n2 = r - m
 
    # create temp arrays
    L = [(0,0)] * (n1)
    R = [(0,0)] * (n2)
 
    # Copy data to temp arrays L[] and R[]
    for i in range(0, n1):
        L[i] = arr[l + i]
 
    for j in range(0, n2):
        R[j] = arr[m + 1 + j]
 
    # Merge the temp arrays back into arr[l..r]
    i = 0     # Initial index of first subarray
    j = 0     # Initial index of second subarray
    k = l     # Initial index of merged subarray
 
    while i < n1 and j < n2:
        if L[i][0] <= R[j][0]:
            arr[k] = L[i]
            i += 1
        else:
            arr[k] = R[j]
            j += 1
        k += 1
 
    # Copy the remaining elements of L[], if there
    # are any
    while i < n1:
        arr[k] = L[i]
        i += 1
        k += 1
 
    # Copy the remaining elements of R[], if there
    # are any
    while j < n2:
        arr[k] = R[j]
        j += 1
        k += 1
def mergeSort(arr, l, r):
    if l < r:
        m = l+(r-l)//2
        mergeSort(arr, l, m)
        mergeSort(arr, m+1, r)
        merge(arr, l, m, r)
```

```python
## Sorting whole cases with respect to the numbers for each person
for i in range(len(wholeCasesList)):
  mergeSort(wholeCasesList[i],0,len(wholeCasesList[i])-1)
```

```python
def findLoongyNumbers(i,j,cache,n):
  if(i==n): #It means it is end
    return 0
  
  if(cache[i][j]!=-1): #Means the value is already in the set
    return cache[i][j]

  if(aCase[i][0]==aCase[j][0]+1):
    v = max(1+findLoongyNumbers(i+1,i,cache,n),findLoongyNumbers(i+1,j,cache,n))
    cache[i][j]=v
    return v
  else:
    v = 0 + findLoongyNumbers(i+1,j,cache,n)
    cache[i][j]=v
    return v
```

```python
aCase = wholeCasesList[0]
n = len(aCase)
cache = [[-1 for j in range(n)] for i in range(n)] #Creating cache for each case
value = findLoongyNumbers(0,0,cache,n)
for i in range(n):
  print(cache[i])
```
